# A Graph Kernel for Dependency Structures

* source: *Collins and Duffy (2001). Parsing with a Single Neuron.
  Convolution Kernels for Natural Language Problems.*
* very similar to the tree kernel described in *Collins and Duffy (2001).
  Convolution Kernels for Natural Language* (and in this technical report as well)

## Dependency graphs

* graph is represented as a set of nodes $\{w_{1},...w_{n}\}$
* $w_{1}$: root node
* all other nodes are ordered **arbitrarily**
* $word(w)$: word at node $w$

* edges signal a dependency from a head-word to a modifier
* edges are labeled with the grammatical relationship between the two
* $children(w)$: set of $(relation, node)$ pairs, which are *direct modifiers* of $w$

## The man saw the woman with the telescope

```python
import networkx as nx
# install with: sudo pip install git+http://github.com/chebee7i/nxpd/#egg=nxpd
from nxpd import draw
from discoursekernels.test_tree import label_nodes, label_edges
```

```python
depgraph = nx.DiGraph()

depgraph.add_nodes_from(label_nodes(
    [(1, '*'), (2, 'saw'), (3, 'man'), (4, 'the'),
     (5, 'woman'), (6, 'the'), (7, 'with'),
     (8, 'telescope'), (9, 'the')]))
depgraph.add_edges_from(label_edges(
    [(1, 2, 'root'), (2, 3, 'sbj'), (3, 4, 'dt'),
     (2, 5, 'obj'), (5, 6, 'dt'),
     (2, 7, 'pp'), (7, 8, 'pp-obj'), (8, 9, 'dt')]
))

draw(depgraph, show='ipynb')
```

* a *feature* is a connected subgraph with at least two words
* a dependency graph $d$ is represented as a vector $\phi(d) = \{h_1(d), h_2(d), ..., h_m(d)\}$
* $h_s(d)$ represents the number of times feature $s$ is seen in graph $d$,
  i.e. how often a specific subgraph occurs in $d$

## Kernel calculation

$$
K(d1, d2) = \sum_{n_1 \in N_1, n_2 \in N_2} Cm(n_1, n_2)
$$

* $N_1$: set of nodes in $d_1$ ...
* $Cm(n_1, n_2)$: number of common subgraphs rooted at $n_1$ and $n_2$

## Common subgraphs calculation

* $sim(n_1, n_2)$: set of common dependencies of $n_1$ and $n_2$

$$
sim(n_1, n_2) = \\
    \{ (x.y) | (Relation, x) \in children(n_1), \\
    (Relation, y \in children(n_2), \\
    word(x) = word(y) \}
$$

* if $word(n_1) \neq word(n_2)$ or $children(n_1) = \emptyset$
  or $children(n_2) = \emptyset$, then $Cm(n_1, n_2) = 0$
* else: $Cm(n_1, n_2) = \prod_{(x,y) \in sim(n_1, n_2)} (Cm(x,y) +2) -1$
