## Dependency graphs

* graph is represented as a set of nodes $\{w_{1},...w_{n}\}$
* $w_{1}$: root node
* all other nodes are **ordered arbitrarily**
* $word(w)$: word at node $w$

* edges signal a dependency from a head-word to a modifier
* edges are labeled with the grammatical relationship between the two
* $children(w)$: set of $(relation, node)$ pairs, which are *direct modifiers* of $w$

```python
def dependency_children(dependency_graph, node, edge_attrib='label'):
    children = set()
    for source, target, edge_attrs in depgraph.out_edges(node, data=True):
        children.add( (edge_attrs[edge_attrib], target) )
    return children
```

```python
dependency_children(depgraph, 2)
```

* a *feature* is a connected subgraph with at least two words
* a dependency graph $d$ is represented as a vector $\phi(d) = \{h_1(d), h_2(d), ..., h_m(d)\}$
* $h_s(d)$ represents the number of times feature $s$ is seen in graph $d$,
  i.e. how often a specific subgraph occurs in $d$

## Kernel calculation

$$
K(d1, d2) = \sum_{n_1 \in N_1, n_2 \in N_2} Cm(n_1, n_2)
$$

* $N_1$: set of nodes in $d_1$ ...
* $Cm(n_1, n_2)$: number of common subgraphs rooted at $n_1$ and $n_2$

## Common subgraphs calculation

* $sim(n_1, n_2)$: set of common dependencies of $n_1$ and $n_2$

$$
sim(n_1, n_2) = \\
    \{ (x.y) | (Relation, x) \in children(n_1), \\
    (Relation, y \in children(n_2), \\
    word(x) = word(y) \}
$$

* if $word(n_1) \neq word(n_2)$ or $children(n_1) = \emptyset$
  or $children(n_2) = \emptyset$, then $Cm(n_1, n_2) = 0$
* else: $Cm(n_1, n_2) = \prod_{(x,y) \in sim(n_1, n_2)} (Cm(x,y) +2) -1$

```python
from collections import defaultdict

# def common_dependencies(graph1, graph2, n1_children, n2_children, node_attrib='label'):
#     common_deps = set()

#     n1_rels = set(rel for rel, target in n1_children)
#     n2_rels = set(rel for rel, target in n2_children)
#     common_rels = n1_relations.intersection(n1_relations)
#     for rel in common_rels:
#         n1_target = ...
        
#     return common_deps

def common_dependencies(graph1, graph2, n1_children, n2_children, node_attrib='label'):
    common_deps = set()

    reldict = defaultdict(lambda : defaultdict(str))
    for rel, target in n1_children:
        reldict[rel]['graph1'] = target
    for rel, target in n2_children:
        reldict[rel]['graph2'] = target
    
    for rel in reldict:
        if 'graph1' in reldict[rel] and 'graph2' in reldict[rel]:
            n1_target = reldict[rel]['graph1']
            n2_target = reldict[rel]['graph2']
            n1_target_word = graph1.node[n1_target][node_attrib]
            n2_target_word = graph2.node[n2_target][node_attrib]
            if n1_target_word == n2_target_word:
                common_deps.add( (n1_target, n2_target) )
    return common_deps

def common_subgraphs(graph1, graph2, n1, n2,
                     node_attrib='label', edge_attrib='label'):
    for graph in (graph1, graph2):
        assert nx.is_directed_acyclic_graph(graph)
    
    if graph1.node[n1][node_attrib] != graph2.node[n2][node_attrib]:
        return 0

    n1_children = dependency_children(graph1, n1, edge_attrib=edge_attrib)
    n2_children = dependency_children(graph2, n2, edge_attrib=edge_attrib)

    if not n1_children or not n2_children:
        return 0
    else:
        result = 1  # neutral element of multiplication
        for n1_target, n2_target in common_dependencies(graph1, graph2,
                                                        n1_children, n2_children,
                                                        node_attrib=node_attrib):
            result *= (common_subgraphs(graph1, graph2,
                                        n1_target, n2_target,
                                        node_attrib='label',
                                        edge_attrib='label') + 2) - 1
        return result
```
