## Dependency graphs

* graph is represented as a set of nodes $\{w_{1},...w_{n}\}$
* $w_{1}$: root node
* all other nodes are **ordered arbitrarily**
* $word(w)$: word at node $w$

* edges signal a dependency from a head-word to a modifier
* edges are labeled with the grammatical relationship between the two
* $children(w)$: set of $(relation, node)$ pairs, which are *direct modifiers* of $w$

* a *feature* is a connected subgraph with at least two words
* a dependency graph $d$ is represented as a vector $\phi(d) = \{h_1(d), h_2(d), ..., h_m(d)\}$
* $h_s(d)$ represents the number of times feature $s$ is seen in graph $d$,
  i.e. how often a specific subgraph occurs in $d$

## Kernel calculation

$$
K(d1, d2) = \sum_{n_1 \in N_1, n_2 \in N_2} Cm(n_1, n_2)
$$

* $N_1$: set of nodes in $d_1$ ...
* $Cm(n_1, n_2)$: number of common subgraphs rooted at $n_1$ and $n_2$

## Common subgraphs calculation

* $sim(n_1, n_2)$: set of common dependencies of $n_1$ and $n_2$

$$
sim(n_1, n_2) = \\
    \{ (x.y) | (Relation, x) \in children(n_1), \\
    (Relation, y \in children(n_2), \\
    word(x) = word(y) \}
$$

* if $word(n_1) \neq word(n_2)$ or $children(n_1) = \emptyset$
  or $children(n_2) = \emptyset$, then $Cm(n_1, n_2) = 0$
* else: $Cm(n_1, n_2) = \prod_{(x,y) \in sim(n_1, n_2)} (Cm(x,y) +2) -1$

```python
import itertools
from networkx.algorithms import isomorphism as iso

def is_dependency_subgraph(graph, subgraph_candidate,
                           node_attrib='label', edge_attrib='label'):
    """
    Collins and Duffy (2001). Parsing with a Single Neuron.::
    
        - a feature of a dependency graph is a connected subgraph
          with at least two words (the root '*' counts as a word, as well)
    """
    if len(subgraph_candidate) > 1:
        if nx.is_weakly_connected(subgraph_candidate):
            if contains_only_complete_dependency_productions(graph, subgraph_candidate,
                                                  node_attrib=node_attrib,
                                                  edge_attrib=edge_attrib):
                return True
    return False

def get_dependency_subgraphs(graph, node_attrib='label', edge_attrib='label'):
    assert nx.is_directed_acyclic_graph(graph)
    for n in xrange(graph.number_of_nodes()):
        for subnodes in itertools.combinations(graph.nodes(), n+1):
            subgraph_candidate = graph.subgraph(subnodes)
            if is_dependency_subgraph(graph, subgraph_candidate,
                                      node_attrib=node_attrib, edge_attrib=edge_attrib):
                yield subgraph_candidate


def generate_all_unique_dependency_subgraphs(graphs, node_attrib='label', edge_attrib='label'):
    same_node_label = iso.categorical_node_match(node_attrib, '')
    same_edge_label = iso.categorical_edge_match(edge_attrib, '')
    if len(graphs) == 0:
        return []
    elif len(graphs) == 1:
        return list(get_dependency_subgraphs(graphs[0], node_attrib=node_attrib))
    else:
        unique_subgraphs = list(get_dependency_subgraphs(
            graphs[0], node_attrib=node_attrib, edge_attrib=edge_attrib))
        for graph in graphs[1:]:
            for subgraph in get_dependency_subgraphs(graph):
                if not any(nx.is_isomorphic(new_subgraph, old_subgraph,
                                            node_match=same_node_label, edge_match=same_edge_label)
                             for new_subgraph, old_subgraph in itertools.product([subgraph], unique_subgraphs)):
                    unique_subgraphs.append(subgraph)
```
